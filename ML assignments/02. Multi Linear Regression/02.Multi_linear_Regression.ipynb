# Example 1

```python
#importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
data = pd.read_csv('assignment/Summary_of_Weather.csv', low_memory=False) 
```

```python
X = np.array(data.MinTemp)
y = np.array(data.MaxTemp)
```

```python
# splitting X and y into training and testing sets 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1) 
X_train= X_train.reshape(-1, 1)
y_train= y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
# create linear regression object 
reg = linear_model.LinearRegression() 

# train the model using the training sets 
reg.fit(X_train, y_train)
```

```python
print('Regression Coefficients: \n', reg.coef_) 
print('Variance score: {}'.format(reg.score(X_test, y_test))) # variance score: 1 means perfect prediction 

## plotting residual errors in training data 
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train, color = "red", s = 10, label = 'Train data') 

## plotting residual errors in test data 
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test, color = "blue", s = 10, label = 'Test data') 
# here reg.predict(X_test) have predicted Y_test value

## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2) 

## plotting legend 
plt.legend(loc = 'upper right') 
plt.title("Residual errors")  
plt.show() 
```

```python
#analysing the effect of salinity of water with temperatre
data = pd.read_csv('assignment/Bottle_salinity.csv', low_memory=False)
X = np.array(data.Salnty)
y = np.array(data.T_degC)
```

```python
# splitting X and y into training and testing sets 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1) 
X_train= X_train.reshape(-1, 1)
y_train= y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
# create linear regression object 
reg = linear_model.LinearRegression() 
reg.fit(X_train, y_train)
print('Regression Coefficients: \n', reg.coef_) 
print('Variance score: {}'.format(reg.score(X_test, y_test))) # variance score: 1 means perfect prediction 

## plotting residual errors in training data 
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train, color = "red", s = 15, label = 'Train data') 

## plotting residual errors in test data 
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test, color = "blue", s = 10, label = 'Test data') 
# here reg.predict(X_test) have predicted Y_test value

## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 30, linewidth = 2) 

## plotting legend 
plt.legend(loc = 'upper right') 
plt.title("Residual errors")  
plt.show() 
```
